```python
import pandas as pd
orders = pd.read_csv("orders.csv")
orders.head()
users = pd.read_json("users.json")
users.head()
import sqlite3

conn = sqlite3.connect("restaurants.db")
cursor = conn.cursor()
with open("restaurants.sql", "r") as f:
    sql_script = f.read()
cursor.execute("DROP TABLE IF EXISTS restaurants;")
conn.commit()

cursor.executescript(sql_script)
conn.commit()
restaurants = pd.read_sql(
    "SELECT * FROM restaurants",
    conn
)

restaurants.head()
orders_users = orders.merge(
    users,
    on="user_id",
    how="left"
)
final_df = orders_users.merge(
    restaurants,
    on="restaurant_id",
    how="left"
)
final_df.head()
final_df.to_csv("final_food_delivery_dataset.csv", index=False)




```

```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")

# Step 1 & 2: total spend per user
user_total_spend = df.groupby('user_id')['total_amount'].sum()

# Step 3: users with spend > 1000
high_value_users = user_total_spend[user_total_spend > 1000]

# Step 4: count distinct users
count_users = high_value_users.count()

count_users
```

```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")
bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ['3.0–3.5', '3.6–4.0', '4.1–4.5', '4.6–5.0']

df['rating_range'] = pd.cut(
    df['rating'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
rating_revenue = (
    df
    .groupby('rating_range')['total_amount']
    .sum()
    .sort_values(ascending=False)
)

rating_revenue
rating_revenue.idxmax()
```

```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")
cuisine_stats = (
    df
    .groupby('cuisine')
    .agg(
        distinct_restaurants=('restaurant_id', 'nunique'),
        total_revenue=('total_amount', 'sum')
    )
    .sort_values('distinct_restaurants')
)

cuisine_stats
```

```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")
restaurant_stats = (
    df
    .groupby(['restaurant_id', 'restaurant_name_x', 'restaurant_name_y'])
    .agg(
        total_orders=('order_id', 'count'),
        avg_order_value=('total_amount', 'mean')
    )
)

low_orders = restaurant_stats[restaurant_stats['total_orders'] < 20] \
                .sort_values('avg_order_value', ascending=False)

low_orders.head(10)

```

```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")

# Ensure order_date is datetime
df['order_date'] = pd.to_datetime(df['order_date'])

# Extract quarter
df['quarter'] = df['order_date'].dt.quarter

# Total revenue per quarter
quarter_revenue = (
    df
    .groupby('quarter')['total_amount']
    .sum()
    .sort_values(ascending=False)
)

quarter_revenue
```
